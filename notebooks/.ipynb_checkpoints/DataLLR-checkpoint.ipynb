```python
def calc_posterior_spline(prior, sigma):
    likelihood = u.gaussian(ut.wallmorphx(x.T),sigma,x)
    post = prior*likelihood
    post = post/post.sum(axis=1,keepdims=True)
    return post
    
    
def cost(params):
    alpha,beta,sigma= params
#     sf,morph,prior = args 
    post = calc_posterior_spline(prior,sigma)
    aff_trans = np.clip(alpha*sf+beta,-1.3,1.3)
    
    morph_dig = np.digitize(wm,x.ravel())-1
    aff_trans_dig = np.digitize(aff_trans,x.ravel())-1
    
    
    prob = post[morph_dig,aff_trans_dig]
    return -np.log(prob).sum() 


# scipy optimize using 
# results = sp.optimize.basinhopping(cost,np.random.rand(3),minimizer_kwargs={'bounds':[[3,None],[None,None],[.1,.6]],'args':[sf,wm,prior]})
```

```python
results = sp.optimize.brute(cost,((2,8),(-1.3,0),(.1,.6)),Ns=10)
```
